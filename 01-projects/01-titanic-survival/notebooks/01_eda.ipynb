```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/raw/titanic.csv")
```

```python
plt.figure(figsize=(6,4))

sns.countplot(data=df, x="Sex", hue="Survived")

plt.title("Survival by Sex")

plt.savefig("../reports/figures/survival_by_sex.png", dpi=200, bbox_inches="tight")

plt.show()
```

## Insight 1

Female passengers had a much higher survival rate than male passengers.
Most male passengers did not survive, while most female passengers survived.

```python
plt.figure(figsize=(6,4))

sns.countplot(data=df, x="Pclass", hue="Survived")

plt.title("Survival by Passenger Class")

plt.savefig("../reports/figures/survival_by_pclass.png", dpi=200, bbox_inches="tight")

plt.show()
```

## Insight 2

Passengers traveling in First Class had a higher survival rate than those in Second and Third Class.
Third Class passengers had the lowest survival rate.

```python
plt.figure(figsize=(8,5))

sns.histplot(data=df, x="Age", hue="Survived", bins=30)

plt.title("Age Distribution by Survival")

plt.savefig("../reports/figures/age_distribution.png",
            dpi=200,
            bbox_inches="tight")

plt.show()
```

# Insights

1. Female passengers had a much higher survival rate than male passengers.

2. Passengers traveling in 1st class survived more often than those in 2nd and 3rd class.

3. Most passengers in 3rd class did not survive.

4. The Age column contains many missing values (177), so it will need imputation before building a machine learning model.

5. The Cabin column has a very large number of missing values (687), so it may need to be removed or handled carefully during preprocessing.
